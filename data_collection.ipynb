```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib as plt
import bs4 as bs
import requests
import datetime
```

```python
#grab a list of tickers and include the general S&P ticker GSPC
resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resp.text,'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})
tickers = ['GSPC']
for row in table.findAll('tr')[1:]:
    ticker=row.findAll('td')[0].text    
    tickers.append(ticker)
```

```python
#clean ticker list
tickers = [ticks.replace('\n', '') for ticks in tickers]
#tickers
```

```python
stock_data = pd.DataFrame()
df_list = []
for ticker in tickers:
    ticker_data = yf.download(ticker)
    ticker_data['Ticker'] = ticker
    df_list = [stock_data,ticker_data]
    stock_data = pd.concat(df_list,ignore_index=False)
```

```python
company_data = pd.DataFrame()
df_list = []
tickers_nonGSPC = tickers[1:]
for ticker in tickers_nonGSPC:
    ticker_data = yf.Ticker(ticker).info
    df= pd.DataFrame.from_dict(ticker_data,orient='index')
    df = df.transpose()
    df['Ticker']=ticker
    df_list = [company_data,df]
    company_data = pd.concat(df_list,ignore_index=False)
```

```python
pd.set_option("display.max_columns", None)
```

```python
#drop columns that do not interest us
simp_company_data = company_data[['Ticker','industry','sector','fullTimeEmployees','totalCash','ebitda','totalDebt','totalRevenue','dividendYield','trailingPE','forwardPE','averageVolume10days','fiftyTwoWeekLow','fiftyTwoWeekHigh','profitMargins','sharesOutstanding','sharesShort','sharesShortPriorMonth','heldPercentInsiders','heldPercentInstitutions','shortRatio','bookValue','priceToBook','trailingEps','forwardEps','recommendationKey','recommendationMean','numberOfAnalystOpinions','returnOnEquity','freeCashflow','operatingCashflow','revenueGrowth','grossMargins','ebitdaMargins','operatingMargins','earningsQuarterlyGrowth','earningsGrowth']]
#simp_company_data
```

```python
#create a DF that indicates the industry 
sector_ind_join = simp_company_data[['Ticker','industry','sector']]
```

```python

stock_data = stock_data.reset_index()
```

```python
#join it to stock data so that I can perform comparison analysis
stock_data_new = pd.merge(stock_data,sector_ind_join,on='Ticker',how='left')
```

```python
stock_data_new
```

```python
#create float column to then be able to perform weighted price comparisons
stock_data_new['Float'] = stock_data_new['Adj Close']*stock_data_new['Volume']
stock_data_new
```

```python
#create industry comparison dataframe
#Float by industry and date
industry_WA = stock_data_new.groupby(['industry','Date']).sum('Float')
#Weighted Average Stock Price
industry_WA['industry_WASP']=industry_WA['Float']/industry_WA['Volume']
industry_WA = industry_WA.reset_index()
industry_WA

```

```python
#create sector comparison dataframe
#Float by sector and date
sector_WA = stock_data_new.groupby(['sector','Date']).sum('Float')
#Weighted Average Stock Price
sector_WA['sector_WASP']=sector_WA['Float']/sector_WA['Volume']
sector_WA = sector_WA.reset_index()
sector_WA
```

```python
#create a copy and merge stock data with industry trends
stock_data_new_copy = stock_data_new
stock_data_comb = pd.merge(stock_data_new,industry_WA[['industry','Date','industry_WASP']],on=['industry','Date'],how='left')
```

```python
#create a copy and merge stock data with sector trends

stock_data_comb = pd.merge(stock_data_comb,sector_WA[['sector','Date','sector_WASP']],on=['sector','Date'],how='left')
```
